```python
import pandas as pd
import re
from IPython.display import display, clear_output

# Function to match headers with columns and fill in the information from user input
def fill_columns_from_input(text):
    columns = ["Project Name", "Project Highlights", "Location Highlights"]
    columns_data = {column: "" for column in columns}

    # Define regular expression pattern to match column headers
    pattern = r"(Project Name|Project Highlights|Location Highlights):(.+?)(?=(Project Name|Project Highlights|Location Highlights|$))"
    matches = re.findall(pattern, text, re.DOTALL)

    # Fill in the information for each match
    for match in matches:
        header, content, _ = match
        columns_data[header.strip()] = content.strip()

    return columns_data

def main():
    text = input("Enter the project details:\n")

    # Clear the output cell
    clear_output(wait=True)

    data = fill_columns_from_input(text)

    # Convert data to Pandas DataFrame
    df = pd.DataFrame([data])  # Create DataFrame from a list containing the data dictionary



    # Display the DataFrame in tabular format
    display(df)

    return df  # Returning df

if __name__ == "__main__":
    df = main()  # Assigning df from the return value of main()
```

```python
import pandas as pd
import re
from IPython.display import display, clear_output

def fill_columns_from_input(text):
    columns = ["Project Name", "Project Highlights", "Location Highlights", "Key Distances",
               "Location Brief (50 words)", "Summary of the Location (100 characters)",
               "List of All Amenities", "Top 5 Highlighted Amenities with Use Cases",
               "Developer Name", "Developer Legacy", "Write-up About the Developer",
               "Short Write-up About the Project", "Configurations and Their Respective Carpet Areas",
               "Tagline Suitable for Online Ad", "Description for Who Should Buy the Project",
               "Website Link and Call Number"]
    columns_data = {column: "" for column in columns}

    pattern = r"(Project Name|Project Highlights|Location Highlights|Key Distances|Location Brief \(50 words\)|Summary of the Location \(100 characters\)|List of All Amenities|Top 5 Highlighted Amenities with Use Cases|Developer Name|Developer Legacy|Write-up About the Developer|Short Write-up About the Project|Configurations and Their Respective Carpet Areas|Tagline Suitable for Online Ad|Description for Who Should Buy the Project|Website Link and Call Number):(.*?)(?=(Project Name|Project Highlights|Location Highlights|Key Distances|Location Brief \(50 words\)|Summary of the Location \(100 characters\)|List of All Amenities|Top 5 Highlighted Amenities with Use Cases|Developer Name|Developer Legacy|Write-up About the Developer|Short Write-up About the Project|Configurations and Their Respective Carpet Areas|Tagline Suitable for Online Ad|Description for Who Should Buy the Project|Website Link and Call Number)|$)"
    matches = re.findall(pattern, text, re.DOTALL)

    for match in matches:
        header, content, _ = match
        columns_data[header.strip()] = content.strip()

    return columns_data

def main():
    projects_data = []

    while True:
        text = input("Enter the project details (or type 'done' to finish):\n")

        if text.lower() == 'done':
            break

        data = fill_columns_from_input(text)
        projects_data.append(data)
        clear_output(wait=True)

    df = pd.DataFrame(projects_data)
    display(df)

    return df

if __name__ == "__main__":
    df = main()
```
